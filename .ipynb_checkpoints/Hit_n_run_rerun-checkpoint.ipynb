# Total incarcerations by year over all charges

```python
df.incar.groupby(df.complaint_year).count()
```

# Incarceration groupings by year and then highest_dispo_charge

This cell counts all incarcerations by highest disposed charged for each year. The column to the far right is the count of cases for that particular charge.

Note: use of groupby ignores null data entries(NaN) in calculations

```python
incar_grp = df['incar'].groupby([df['complaint_year'], df['highest_dispo_charge']])
incar_grp.count()
```

# Average incarceration time by year for each highest charge disposed

```python
incar_grp.mean()
```

# List by year of the case and incarceration time given

The cell below lists by year the index for the case (far left on the data set window at bottom) and how many days of incarceration given for that case.
*note: left column are indexes in the data frame

```python
for (a,b), c in incar_grp :
    print a,b
    print c
```

```python
no_incar_cases = df[df.apply( lambda x: pd.isnull(x['incar']) == True and pd.isnull(x['final_disposition']) == False, axis=1)]
```

```python
no_incar_cases  #these still have dismissed cases!!!
```

```python
#need to get rid of dismissed cases
no_incar_cases_minus_dismissed = no_incar_cases[~no_incar_cases.final_disposition.str.contains("Dismissed")]

       
```

# Total number of cases without incarceration

We have 212 total without incarceration

```python
len(no_incar_cases_minus_dismissed)  #Total cases with no incarceration...entire data set
```

# Totals by year for cases without incarceration

```python
noincar_by_year_grp = no_incar_cases_minus_dismissed.groupby( no_incar_cases['complaint_year'])
```

```python
noincar_by_year_grp.complaint_year.count()   #These are the by year counts for non-null final disposition and no incarceration
                                             #basically these are the people who didn't get incarcerated by year
    
```
